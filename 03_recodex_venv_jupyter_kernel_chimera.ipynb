# Virtual environment (venv)

- A virtual environment is a self-contained directory that contains a Python installation and its own set of packages. It allows you to manage dependencies for different projects separately, avoiding conflicts between them.

- Virtual environments are useful when you have multiple projects that require different versions of Python or some packages, or when you want to avoid installing packages globally on your system.

- It is recommended to use virtual environments for your Python projects! For example, you can create a virtual environment for our course and install the required packages there.

- In Windows, you can create a virtual environment using the following command in the cmd terminal:

```
py -3.13 -m venv venv-nofy084
```

- Creating it isn't enough; you have to "enter" it. On Windows, the activation script is located in the Scripts folder.

In PowerShell (Standard):

```
.\venv-nofy084\Scripts\Activate.ps1
```

In Command Prompt (CMD):

```
.\venv-nofy084\Scripts\activate.bat
```

If you are using PowerShell and get an error saying "script execution is disabled on this system," Windows is just being protective. You can fix this for the current session by running:

```
Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope Process
```

# Jupyter kernel

- Jupyter notebooks are a popular tool for interactive programming and data analysis. They allow you to write and execute code in a web-based interface, and they support multiple programming languages through the use of kernels.

- To use a virtual environment in Jupyter notebooks, you need to install the ipykernel package in your virtual environment, first. Activate the virtual environment and then run the following command in the terminal:

```
pip install ipykernel
```

- Once ipykernel is installed, VS Code can do the rest for you. When you open a Jupyter notebook in VS Code, click "Select Kernel" in the top-right corner. Choose "Python Environments..." and your `venv-nofy084` should appear in the list. If you haven't install the ipykernel package in the virtual environment you are selecting, VS Code will offer you installation of the package, when you try to run a Jupyter Notebook:
