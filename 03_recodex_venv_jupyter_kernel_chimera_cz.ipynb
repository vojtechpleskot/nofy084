# Virtuln prosted (venv)

- Virtuln prosted je samostatn adres, kter obsahuje instalaci Pythonu a vlastn sadu balk. Umouje vm spravovat zvislosti rznch projekt oddlen, m se vyhnete konfliktm mezi nimi.

- Virtuln prosted jsou uiten, kdy mte vce projekt, kter vyaduj rzn verze Pythonu nebo nkterch balk, nebo kdy chcete vyhnout se globln instalaci balk do celho systmu.

- Je doporueno pouvat virtuln prosted pro vae Python projekty! Mete napklad vytvoit virtuln prosted pro n kurz a tam nainstalovat poadovan balky.

- Ve Windows mete vytvoit virtuln prosted pomoc nsledujcho pkazu v cmd terminlu:

```
python -m venv venv-nofy084
```

- Pouh vytvoen venv nesta; muste do nj "vstoupit". Ve Windows se aktivan skript nachz ve sloce Scripts.

V PowerShell (Standard):

```
.\venv-nofy084\Scripts\Activate.ps1
```

V Command Prompt (CMD):

```
.\venv-nofy084\Scripts\activate.bat
```

# Jupyter kernel

- Jupyter notebooky jsou oblben nstroj pro interaktivn programovn a analzu dat. Umouj vm pst a spoutt kd ve webovm rozhran a podporuj vce programovacch jazyk prostednictvm kernel.

- Pro pouit virtulnho prosted v Jupyter noteboocch muste nejprve nainstalovat balek ipykernel ve vaem virtulnm prosted. Aktivujte virtuln prosted a pot spuste nsledujc pkaz v terminlu:

```
pip install ipykernel
```

- Jakmile je ipykernel nainstalovn, VS Code me udlat zbytek za vs. Kdy otevete Jupyter notebook ve VS Code, kliknte na "Select Kernel" v pravm hornm rohu. Vyberte "Python Environments..." a v `venv-nofy084` by se ml objevit v seznamu.
